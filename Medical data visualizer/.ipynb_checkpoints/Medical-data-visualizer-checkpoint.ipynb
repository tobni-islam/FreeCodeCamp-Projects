```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Import data
df = pd.read_csv("medical_examination.csv")
df.head()
```

```python
# Add 'overweight' column (If that value of BMI is > 25 then the person is overweight)
df['overweight'] = (df["weight"] / (df["height"] / 100) ** 2).apply(lambda x: 1 if x>25 else 0)

df.head()
```

```python
# Normalize data by making 0 always good and 1 always bad. 
#If the value of 'cholesterol' or 'gluc' is 1, make the value 0. If the value is more than 1, make the value 1.

df["cholesterol"] = df["cholesterol"].apply(lambda x: 0 if x == 1 else 1)
df["gluc"] = df["gluc"].apply(lambda x: 0 if x ==1 else 1)

df.head()
```

# Draw Categorical Plot

```python
# Create DataFrame for cat plot just the values from 'cholesterol', 'gluc', 'smoke', 'alco', 'active', and 'overweight'
df_cat = df.melt(id_vars = ["cardio"],value_vars = ["active","alco","cholesterol","gluc","overweight","smoke"])

# Group and reformat the data to split it by 'cardio'.
df_cat = ((df_cat.groupby(["variable","cardio"]).value_counts()).rename("total")).reset_index()

df_cat.head()
```

```python
# Draw the catplot
g = sns.catplot(data = df_cat,x = "variable",y = "total",kind = "bar",hue = "value",col = "cardio")

plt.show()
```

```python
# Get the figure for the output and save it
fig = g.fig
fig.savefig('catplot.png')
```

# Draw Heat Map

```python
# Clean the data
df_heat = df[(df["ap_lo"] <= df["ap_hi"]) & (df["height"] >= df["height"].quantile(0.025)) & (df["height"] <= df["height"].quantile(0.975)) \
             & (df["weight"] >= df["weight"].quantile(0.025)) & (df["weight"] <= df["weight"].quantile(0.975))]

df_heat.head()
```

```python
# Calculate the correlation matrix
corr = df_heat.corr()
corr
```

```python
# Generate a mask for the upper triangle
mask = np.triu(corr)
```

```python
# Set up the matplotlib figure
fig, ax = plt.subplots()

# Draw the heatmap with 'sns.heatmap()'
sns.heatmap(corr,mask = mask,annot =True,fmt = ".1f",linewidths = 0.5,cbar_kws = {"shrink":0.6},cmap = "icefire",vmin = -0.5,ax=ax)

plt.show()
```

```python
# Save the figure
fig.savefig('heatmap.png')
```

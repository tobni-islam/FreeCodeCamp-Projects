```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()
```

```python
# Import data
df = pd.read_csv("fcc-forum-pageviews.csv",index_col = "date",parse_dates = ["date"])

df.head()
```

```python
# Clean data
df = df[(df["value"] >= df["value"].quantile(0.025)) & (df["value"] <= df["value"].quantile(0.975))]

df.head()
```

# Draw line plot

```python
plt.figure(figsize=(15, 7))
plt.plot(df.index, df["value"], color="red")
plt.title("Daily freeCodeCamp Forum Page Views 5/2016-12/2019")
plt.xlabel("Date")
plt.ylabel("Page Views")

plt.show()
```

```python
# Get the figure and save it
fig = plt.gcf()
fig.savefig('line_plot.png')
```

# Draw bar plot

```python
# Copy and modify data for monthly bar plot
df_bar = df.copy()
df_bar["year"] = df.index.year
df_bar["month"] = df.index.month_name()
df_bar = df_bar.groupby(["year", "month"])["value"].mean().reset_index()

df_bar.head()
```

```python
# Set colors order
colors = ["blue", "orange", "green", "red", "purple", "brown", "pink", "grey", "green", "skyblue", "blue", "orange"]

# Set months order
order_months = ["January", "February", "March", "April", "May", "June", "July", "August", "September", "October", "November",
         "December"]

# Draw bar plot
plt.figure(figsize=(10, 10))
sns.barplot(data=df_bar, x="year", y="value", hue="month", hue_order=order_months, palette=colors)
plt.xlabel("Years")
plt.ylabel("Average Page Views")

plt.legend()
plt.show()
```

```python
# Get the figure and save it
fig = plt.gcf()
fig.savefig('bar_plot.png')
```

# Draw box plots

```python
# Prepare data for box plots 
df_box = df.copy()
df_box.reset_index(inplace=True)
df_box['year'] = [d.year for d in df_box.date]
df_box['month'] = [d.strftime('%b') for d in df_box.date]

df_box.head()
```

```python
# Set the months order
order_months = ['Jan','Feb','Mar','Apr','May','Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
```

```python
# Draw tow box plots
fig, ax = plt.subplots(1, 2, gridspec_kw={"wspace": 0.2}, figsize=(20, 7))

sns.boxplot(data=df_box, x="year", y="value", ax=ax[0])
ax[0].set_title("Year-wise Box Plot (Trend)")
ax[0].set_xlabel("Year")
ax[0].set_ylabel("Page Views")

sns.boxplot(data=df_box, x="month", y="value", ax=ax[1],order=order_months)
ax[1].set_title("Month-wise Box Plot (Seasonality)")
ax[1].set_xlabel("Month")
ax[1].set_ylabel("Page Views")

plt.show()
```

```python
# Get the figure and save it
fig = plt.gcf()
fig.savefig('box_plot.png')
```
